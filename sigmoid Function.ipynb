## Sigmoid Function
A sigmoid function is a mathematical function having a characteristic "S"-shaped curve or sigmoid curve. A standard choice for a sigmoid function is the logistic function 

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Define sigmoid function
def sigmoid(x,w,b):
    return (1.0)/(1.0 +(np.exp(-(w*x +b))))
```

### Finally implementing the Sigmoid function that will update w and b using gradient vector

#### Our sample data looks like this

```python
x = np.array([-2,-1,0,2,3,4,5,6,7])
y = np.array([0.1978, 0.3543, 0.5489, 0.8581,0.9308, 0.9677,0.9852, 0.9933, 0.9970])

plt.xlim(-10,10)
plt.ylim(0,1.2)
plt.plot(x,y,'*')
plt.title("Plot of Data")
plt.show()
```

```python
class SigmoidFunction:
    
    def __init__(self):
        self.w = None
        self.b = None
        
    def sigmoid(self,x):
        return (1.0/(1.0 + np.exp(-(np.dot(x,self.w) +self.b))))
    
    def Loss(self, X,Y):
        loss = 0
        for x,y in zip(X,Y):
            loss += (y - self.sigmoid(x))**2
        return loss/9 #We have 9 inputs in our data
        
    def grad_w(self, x,y):
        y_est = self.sigmoid(x)
        return 2*(y_est - y) * y_est * (1-y_est) * x
    
    def grad_b(self,x,y):
        y_est = self.sigmoid(x)
        return 2*(y_est - y) * y_est * (1- y_est)
    
    def fit(self, X,Y,eta=1, initialize = True):
        if initialize:
            self.w = np.random.ranf()
            self.b = 0
        dw = 0
        db = 0
        
        for x,y in zip(X,Y):
            dw += self.grad_w(x,y)
            db += self.grad_b(x,y)
        
        self.w -= eta * dw
        self.b -= eta * db
        
        
```

```python
#Test Run
sf = SigmoidFunction()
sf.fit(x,y,0.5,True)
```

```python
#Printing w and b after 1 iteration
sf.w, sf.b,sf.Loss(x,y)
```

```python
sf.fit(x,y,0.5,False)
sf.w, sf.b,sf.Loss(x,y)
```

```python
#Lets run for some 30 iterations and visulize how our data is beign fited
X = np.array([-2,-1,0,2,3,4,5,6,7])
Y = np.array([0.1978, 0.3543, 0.5489, 0.8581,0.9308, 0.9677,0.9852, 0.9933, 0.9970])

for x in range(0,30):
    if x == 0:
        sf.fit(X,Y,0.5,True)
    else:
        sf.fit(X,Y,0.5,False)
    #Plotting the function with each iteration
    xx = np.linspace(-10,10,100)
    yy = sf.sigmoid(xx)
    plt.plot(xx,yy)
    plt.plot(X,Y,'*') #Plotting the data points to compare with original points
    plt.text(0,0,'w = {}, b = {}, Loss = {} '.format(sf.w, sf.b, sf.Loss(X,Y)))
    plt.xlim(-10,10)
    plt.ylim(0,1.2)
    plt.show()
```

```python
#Running again
for x in range(0,30):
    if x == 0:
        sf.fit(X,Y,0.5,True)
    else:
        sf.fit(X,Y,0.5,False)
    #Plotting the function with each iteration
    xx = np.linspace(-10,10,100)
    yy = sf.sigmoid(xx)
    plt.plot(xx,yy)
    plt.plot(X,Y,'*') #Plotting the data points to compare with original points
    plt.text(0,0,'w = {}, b = {}, Loss = {} '.format(sf.w, sf.b, sf.Loss(X,Y)))
    plt.xlim(-10,10)
    plt.ylim(0,1.2)
    plt.show()
```

```python
#Now lets run the loop untill its satisfied
sf.fit(X,Y,0.3,True) # I have reduced the value of ets(learning rate)
counter = 1
Loss = sf.Loss(X,Y)
print("Initial Loss = ",Loss)
while Loss > 0.00001:
    sf.fit(X,Y,0.3,False)
    Loss = sf.Loss(X,Y)
    counter += 1

print('Total number of Run = ',counter)
print('w = ', sf.w)
print('b = ', sf.b)
print('Final Loss = ', sf.Loss(X,Y))
```
