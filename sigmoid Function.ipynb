## Sigmoid Function
A sigmoid function is a mathematical function having a characteristic "S"-shaped curve or sigmoid curve. A standard choice for a sigmoid function is the logistic function 

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Define sigmoid function
def sigmoid(x,w,b):
    return (1.0)/(1.0 +(np.exp(-(w*x +b))))
```

```python
#Testing
sigmoid(5,-1,2)
```

```python
w= 2
b= 1.5
x= np.linspace(-10,10,100)
y = sigmoid(x,w,b)
plt.plot(x,y)
plt.title("Plot of Sigmoid function")
plt.show()
```

### Conclusion

#####  We have seen  as how the sigmoid function varies as we change the values of parameters w and b.

- When w is positive : The plot will be at zero and then increase to 1 as we increase the value of x

- when w is negative : The plot flips, The plot will be 1 initially, when we increse the value of X, the plot values decrease to zero.

- when we increase the value of w, the region of change decreases, and this happens the same when we decrease the value of w.

- when we increase the value of b, the point of drop will shift towards right.

- whne we increase the value of b, the point of drop will shift towards right.
